```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/archive (4).zip")
```

```python
def missing_data(x):
    return (x.isnull().sum() / len(x)) * 100
```

```python
missing_data(df)
```

```python
sns.countplot(data = df, x = "Cell Line")
```

```python
sns.displot(data = df, x = "logFC", kind = "kde", hue = "Cell Line")
```

```python
g = sns.FacetGrid(df, col="Cell Line")
g.map_dataframe(sns.histplot, x="P.Value")
```
